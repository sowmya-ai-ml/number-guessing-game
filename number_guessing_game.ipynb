```python
import random
high_score = 0
name = input("Enter your name:")
print("Welcome to number guessing game!",name)
print("Select difficulty level:")
print("1. Easy (1 - 50)")
print("2. Medium (1 - 100)")
print("3. Difficult (1 - 200)")
choice = int (input("Enter your choice:"))
if choice == 1:
  number = random.randint(1,50)
  attempts = 10
elif choice == 2:
   number = random.randint(1,100)
   attempts = 7
else :
   number = random.randint(1,200)
   attempts = 5
print(name,"you have",attempts,"attempts to guess the number")
while attempts >0:
  guess = int(input("Enter your guess:"))
  if guess == number:
      score = attempts *10
      print("Congratulations!",name,"You guessed the number!")
      print("Your score :",score)
      if score > high_score:
        high_score = score
        print("New High Score!")
      print("High score:",high_score)
      break
  elif guess < number:
      print("Too low!")
  else:
      print("Too high!")
  attempts -= 1
  print("Attempts left:",attempts)
if attempts == 0:

   print("Game over!",name," The number was:",number)
   print("Good luck",name,"!")
play_again = input("Do you want to play again?(yes/no):")
if play_again == "yes":
     print("Restart the program")
else:
     print("Thanks for playing")
```
